```python
whisper.available_models()
```

Create a model directory for model storage

```python
MODEL_DIR = os.path.join(os.getcwd(), "../models")
os.makedirs(MODEL_DIR, exist_ok=True)
```

```python
baseModel = whisper.load_model("base",download_root=MODEL_DIR)
```

```python
mediumModel = whisper.load_model("medium",download_root=MODEL_DIR)
```

```python
large_v1Model = whisper.load_model("large-v1",download_root=MODEL_DIR)
```

```python
large_v3Model = whisper.load_model("large-v3",download_root=MODEL_DIR)
```

```python
large_v3_turboModel = whisper.load_model("large-v3-turbo",download_root=MODEL_DIR)
```

```python
result = baseModel.transcribe("../audios/EN_today_date.m4a")
result
```

```python
# Base
result = baseModel.transcribe("../audios/CH_Where_is_toronto.m4a",language = "Zh",verbose=True )
result
```

```python
# Medium
result = mediumModel.transcribe("../audios/CH_Where_is_toronto.m4a",language = "Zh",verbose=True )
result
```

```python
# V1
result = large_v1Model.transcribe("../audios/CH_Where_is_toronto.m4a",language = "Zh",verbose=True )
result['text']
```
